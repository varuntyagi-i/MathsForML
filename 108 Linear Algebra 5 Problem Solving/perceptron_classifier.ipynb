```python
def plot_line(w,wo,x,y):
    x_val = np.linspace(-4,16,100)
    y_val =  (-wo-(w[0]*x_val))/w[1]
    sns.scatterplot(data=df, x ='x1', y='x2', hue='y', palette=['g','r'])
    plt.plot(x_val, y_val, c='b')
    plt.show()
plot_line(w,wo,x,y)
```

```python
def distance_from_line(w,wo,x):
    d = (np.dot(w,x)+wo)/np.linalg.norm(w)
    return d
```

```python
distance_from_line(w,wo,[10,5])
```

```python
def classifier(w,wo,x,y,iters=100):
    for i in range(iters):
        num_of_misclassification = 0
        for xi,yi in zip(x,y):
            dist = distance_from_line(w,wo,xi)
            hs = np.sign(dist)
            if hs != yi:
                num_of_misclassification += 1
                w = w+(xi*yi)
                wo = wo+yi 
        print("Accuracy: {}".format(100-(100*num_of_misclassification/len(y))))
        print("Weight: wo = {}, w1 = {}, w2 = {}".format(wo,w[0],w[1]))
        plot_line(w,wo,x,y)
        if num_of_misclassification == 0:
            return w,wo 
    return w,wo

classifier(w,wo,x,y)
```
