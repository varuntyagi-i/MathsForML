```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def gradient(func,x,y):
    delta = 0.0001
    fx = (func(x+delta, y) - func(x, y)) / delta
    fy = (func(x, y+delta) - func(x, y)) / delta

    return fx, fy 
```

```python
axis_lims = 10
x = np.linspace(-axis_lims, axis_lims, 21)
y = np.linspace(-axis_lims, axis_lims, 21)

# Below grid will have 2 array, and each point of both array we will join to create a coordinate.
g1,g2 = np.meshgrid(x,y) 
g1, g2
```

```python
plt.plot(g1,g2,".y") # b is color here
plt.show()
```

```python
def get_grid(axis_lims):
    x = np.linspace(-axis_lims, axis_lims, 21)
    y = np.linspace(-axis_lims, axis_lims, 21)
    
    g1, g2 = np.meshgrid(x,y)

    return g1, g2
```

```python
def get_gradient(func, g1,g2):
    u = np.zeros(g1.shape)
    v = np.zeros(g2.shape)
    for row_id, (row_x, row_y) in enumerate(zip(g1,g2)):
        for col_id, (x,y) in enumerate(zip(row_x, row_y)):
            fx, fy = gradient(func, x, y)
            u[row_id, col_id] = fx
            v[row_id, col_id] = fy
    
    return u,v 
```

```python
def plot_gradient(g1, g2, u, v):
    plt.quiver(g1, g2, u, v)
    plt.show()
```

```python
func = lambda x,y:x**2+y**2
```

```python
axis_lims = 10
g1,g2 = get_grid(axis_lims)
u, v = get_gradient(func, g1,g2)
plot_gradient(g1, g2, u, v)
```

Points from above graph
1. Bigger vectors at extreme points on the coordinates.
2. Bigger vector means my slope is bigger or more steeper on extreme points.
3. Also all my vectors are pointing outwards, that means all are in direction away from origin.

```python
axis_lims = 10
g1,g2 = get_grid(axis_lims)
u, v = get_gradient(func, g1,g2)
plot_gradient(g1, g2, -u, -v)

# so if we move away in opposite/negative direction of gradient we will reach the minimum of the function. 
```
